# Linear Classifiers (logistic regression and GDA)

#### Dataset 1

```python
x_train, y_train = util.load_dataset('../data/ds1_train.csv', add_intercept=False)
```

```python
# let us visualize the training data set
# data looks like could benefit from linear classifier
# let us compare logistic and GDA
# weak vs. strong assumptions
plt.figure()
plt.plot(x_train[y_train == 1, 0], x_train[y_train == 1, 1], 'bo', linewidth=2)
plt.plot(x_train[y_train == 0, 0], x_train[y_train == 0, 1], 'go', linewidth=2)
plt.title("training data")
```

```python
# might be easier to see
color = ['blue' if y == 1 else 'green' for y in y_train]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_train[:,0], x_train[:,1], y_train, c = color)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.view_init(20, -120)
```

```python
# might be easier to see
color = ['blue' if y == 1 else 'green' for y in y_train]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_train[:,0], x_train[:,1], y_train, c = color)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.view_init(90, 360)
```

```python
# let us look at the distribution of the features by label
# label = 1
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

axs[0].hist(x_train[y_train == 1,0])
axs[0].set_title("Label 1, Features x1")
axs[1].hist(x_train[y_train == 1,1])
axs[1].set_title("Label 1, Features x2")
```

```python
# label = 0
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

axs[0].hist(x_train[y_train == 0,0])
axs[0].set_title("Label 0, Features x1")
axs[1].hist(x_train[y_train == 0,1])
axs[1].set_title("Label 0, Features x2")
```

GDA will not perform too well here, because GDA makes the assumption of normality for all features. It looks like $X_2$ is distributed lognormally.

#### Dataset 2

```python
x_train, y_train = util.load_dataset('../data/ds2_train.csv', add_intercept=False)
```

```python
plt.figure()
plt.plot(x_train[y_train == 1, 0], x_train[y_train == 1, 1], 'bo', linewidth=2)
plt.plot(x_train[y_train == 0, 0], x_train[y_train == 0, 1], 'go', linewidth=2)
plt.title("training data")
```

```python
# might be easier to see
color = ['blue' if y == 1 else 'green' for y in y_train]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_train[:,0], x_train[:,1], y_train, c = color)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.view_init(90, 360)
```

```python
# let us look at the distribution of the features by label
# label = 1
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

axs[0].hist(x_train[y_train == 1,0])
axs[0].set_title("Label 1, Features x1")
axs[1].hist(x_train[y_train == 1,1])
axs[1].set_title("Label 1, Features x2")
```

```python
# label = 0
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

axs[0].hist(x_train[y_train == 0,0])
axs[0].set_title("Label 0, Features x1")
axs[1].hist(x_train[y_train == 0,1])
axs[1].set_title("Label 0, Features x2")
```

```python
# Let us look at the other assumptions needed for LDA 
# (essentially shared covariance matrix for classes)
x = x_train.copy()
y= y_train.copy()


mu_1 = x[y == 1, :].sum(axis=-2) / (y == 1).sum()
mu_0 = x[y == 0, :].sum(axis=-2) / (y == 0).sum()

cov1 = x[y == 1, :]
cov1 -= mu_1
cov1 = cov1.T.dot(cov1) / cov1.shape[0]

cov0 = x[y == 0, :]
cov0 -= mu_0
cov0 = cov0.T.dot(cov0) / cov0.shape[0]

print(cov0)
print(mu_0)
print("\n")
print(cov1)
print(mu_1)
```

GDA on dataset 2 should, in theory, perform better.
